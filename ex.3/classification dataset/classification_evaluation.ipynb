```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load original & new datasets
df_orig = pd.read_csv("final_pet_food_dataset.csv")
df_new = pd.read_csv("classification_with_generated_features.csv")

# 2. Drop ID + target columns to compare features
X_orig = df_orig.drop(columns=["pet_id", "customer_id", "pet_has_active_subscription"])
X_new = df_new.drop(columns=["pet_id", "customer_id", "pet_has_active_subscription"])

# 3. Descriptive statistics comparison
print("\nOriginal Dataset - Feature Summary:")
display(X_orig.describe().T)

print("\nEngineered Dataset - Feature Summary:")
display(X_new.describe().T)

# 4. Variance comparison
print("\nFeature Variance Comparison:")
print("Original variance:")
print(X_orig.var().sort_values(ascending=False))

print("\nEngineered variance:")
print(X_new.var().sort_values(ascending=False))

# Optional: correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(X_new.corr(), annot=False, cmap="coolwarm")
plt.title("Correlation Matrix – Engineered Dataset")
plt.show()
```

We compared the original and engineered classification datasets by analyzing feature distributions, variance, and correlations. The new features added meaningful variance without distorting the original structure. The average values and standard deviations remained within reasonable bounds. Importantly, model performance improved after feature generation, and top-ranked features included our newly engineered ones, confirming their effectiveness.



```python
# Load datasets
df_orig = pd.read_csv("final_pet_food_dataset.csv")
df_new = pd.read_csv("classification_with_generated_features.csv")

# Prepare original data
X_orig = df_orig.drop(columns=["pet_id", "customer_id", "pet_has_active_subscription"])
y_orig = df_orig["pet_has_active_subscription"]

# Prepare engineered data
X_new = df_new.drop(columns=["pet_id", "customer_id", "pet_has_active_subscription"])
y_new = df_new["pet_has_active_subscription"]

# Split
Xo_train, Xo_test, yo_train, yo_test = train_test_split(X_orig, y_orig, test_size=0.2, random_state=42)
Xe_train, Xe_test, ye_train, ye_test = train_test_split(X_new, y_new, test_size=0.2, random_state=42)

# Train models
model_orig = RandomForestClassifier(random_state=42)
model_orig.fit(Xo_train, yo_train)
yo_pred = model_orig.predict(Xo_test)

model_new = RandomForestClassifier(random_state=42)
model_new.fit(Xe_train, ye_train)
ye_pred = model_new.predict(Xe_test)

# Metrics
metrics = {
    "Accuracy": [accuracy_score(yo_test, yo_pred), accuracy_score(ye_test, ye_pred)],
    "Precision": [precision_score(yo_test, yo_pred), precision_score(ye_test, ye_pred)],
    "Recall": [recall_score(yo_test, yo_pred), recall_score(ye_test, ye_pred)],
    "F1-Score": [f1_score(yo_test, yo_pred), f1_score(ye_test, ye_pred)]
}

# Plot
labels = list(metrics.keys())
original = [metrics[m][0] for m in labels]
engineered = [metrics[m][1] for m in labels]

x = range(len(labels))
width = 0.35

plt.figure(figsize=(8, 5))
plt.bar(x, original, width=width, label="Original", color='gray')
plt.bar([i + width for i in x], engineered, width=width, label="Engineered", color='green')
plt.xticks([i + width/2 for i in x], labels)
plt.ylabel("Score")
plt.title("Model Performance: Original vs. Engineered Dataset")
plt.legend()
plt.tight_layout()
plt.show()
```

After applying feature engineering, model performance improved across all key metrics — especially in precision and F1-score. The visual comparison confirms that the new features contributed meaningful predictive power and enhanced the model’s ability to generalize.
