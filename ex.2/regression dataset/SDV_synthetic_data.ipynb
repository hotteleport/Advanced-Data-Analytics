```python
from sdv.lite import TabularPreset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("ship_fuel_efficiency_numeric.csv")

# Fit SDV model
model = TabularPreset(name="FAST_ML")
model.fit(df)

# Generate synthetic data
synthetic_data = model.sample(1000)

# Plot comparison of fuel_consumption
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
sns.histplot(df["fuel_consumption"], bins=40, kde=True)
plt.title("Original fuel_consumption")

plt.subplot(1, 2, 2)
sns.histplot(synthetic_data["fuel_consumption"], bins=40, kde=True, color="orange")
plt.title("Synthetic fuel_consumption")

plt.tight_layout()
plt.show()
```

We used SDV’s GaussianCopula model to generate 1,000 new synthetic rows based on the original regression dataset, successfully preserving the distribution of key variables like fuel_consumption.
